```python
import numpy as np
a1 = [250, 200]
a2 = [50, 1000]
dice = np.array([a1, a2])
```

```python
from scipy import stats

stats.chi2_contingency(dice)
```

```python
chi2_stat, p_val, dof, ex = stats.chi2_contingency(dice)
print("Chi2 Stat")
print(chi2_stat)
print("\n")
print("Degrees of Freedom")
print(dof)
print("\n")
print("Expected value table")
print(ex)
```
